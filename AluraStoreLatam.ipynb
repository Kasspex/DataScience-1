### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Ingreso total de cada tienda
total_tienda = tienda["Precio"].sum()
total_tienda2 = tienda2["Precio"].sum()
total_tienda3 = tienda3["Precio"].sum()
total_tienda4 = tienda4["Precio"].sum()
```

# 2. Ventas por categoría

```python
#Categorias mas populares
maspopular = tienda["Categoría del Producto"].value_counts()
maspopular2 = tienda2["Categoría del Producto"].value_counts()
maspopular3 = tienda3["Categoría del Producto"].value_counts()
maspopular4 = tienda4["Categoría del Producto"].value_counts()
```

# 3. Calificación promedio de la tienda


```python
#Valoracion promedio
calificacion = tienda['Calificación'].mean()
calificacion2 = tienda2['Calificación'].mean()
calificacion3 = tienda3['Calificación'].mean()
calificacion4 = tienda4['Calificación'].mean()
```

# 4. Productos más y menos vendidos

```python
masvendido = tienda['Producto'].value_counts()
masvendido2 = tienda2['Producto'].value_counts()
masvendido3 = tienda3['Producto'].value_counts()
masvendido4 = tienda4['Producto'].value_counts()

masvendido4.head()
```

```python
menosvendido = tienda['Producto'].value_counts(ascending=True)
menosvendido2 = tienda2['Producto'].value_counts(ascending=True)
menosvendido3 = tienda3['Producto'].value_counts(ascending=True)
menosvendido4 = tienda4['Producto'].value_counts(ascending=True)
```

# 5. Envío promedio por tienda

```python
envio = tienda['Costo de envío'].mean()
envio2 = tienda2['Costo de envío'].mean()
envio3 = tienda3['Costo de envío'].mean()
envio4 = tienda4['Costo de envío'].mean()
```

```python
import matplotlib.pyplot as plt

promedios = [calificacion, calificacion2, calificacion3, calificacion4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.bar(tiendas, promedios)
plt.ylim(3.95, 4.06)

plt.title('Calificación promedio de las tiendas')
plt.ylabel('Calificación')
plt.xlabel('Tiendas')


plt.show()
```

```python
import matplotlib.ticker as mticker

total = [total_tienda, total_tienda2, total_tienda3, total_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

fig, ax = plt.subplots()
ax.barh(tiendas, total, color='skyblue')

# 3. Formateador para que los números del eje X sean legibles
# Le decimos que divida cada número entre 1 millón y añada una " M"
formatter = mticker.FuncFormatter(lambda x, pos: f'{x/1e6:,.0f} M')
ax.xaxis.set_major_formatter(formatter)


# 4. Etiquetas y título correctos
ax.set_title('Ingreso total de cada tienda')
ax.set_xlabel('Ingreso ($)')
ax.set_ylabel('Tiendas')

# 5. Mostrar el gráfico
plt.show()
```

```python
envioTotal = [envio, envio2, envio3, envio4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.pie(envioTotal, labels=tiendas, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Envio promedio por tienda')
plt.show()
```
